```python
import os
import glob
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
```

```python
best_exp = './ZBTB7A-multiple_exp_adam_005_200_epoch/test/8.txt'
save_path = './ZBTB7A-multiple_exp_adam_005_200_epoch/test/'
```

```python
f = open(best_exp, 'r')
data = f.read()
f.close()
```

```python
res = data.split('\n')
```

```python
train_loss = []
train_auc = []

test_loss = []
test_auc = []
```

```python
for line_idx in range(len(res)-1):
    train = res[line_idx].split("|")[0]
    test = res[line_idx].split("|")[1]

    train_auc.append(float(train.split(":")[0]))
    train_loss.append(float(train.split(":")[1].split("+")[0]) + float(train.split(":")[1].split("+")[1]))

    test_auc.append(float(test.split(":")[0]))
    test_loss.append(float(test.split(":")[1].split("+")[0]) + float(test.split(":")[1].split("+")[1]))
```

```python
plt.figure(figsize = (6, 10))
plt.subplot(2,1,1)
plt.plot(train_auc, label = 'Train AUC')
plt.plot(test_auc, label = 'Test AUC')
plt.ylim(0.3, 1.0)
plt.legend()
plt.subplot(2,1,2)
plt.plot(train_loss, label = 'Train Loss')
plt.plot(test_loss, label = 'Test Loss')
plt.yscale("log")
plt.legend()
# plt.savefig(save_path + 'best.png')
```
