```python
import os
import pandas as pd
import matplotlib.pyplot as plt

path = './'
filelist = os.listdir(path)
```

```python
filelist[1]
```

```python
trainAUC1 = []
trainAUC2 = []
trainLoss = []
validAUC1 = []
validAUC2 = []
validLoss = []

f = open(path+filelist[1], 'r')
line = f.readline()
while True:
    line = f.readline()
    if '(' in line :
        parsed = line.split('\n')[0].split('(')[1].split(')')[0].split(',')
        trainAUC1.append(float(parsed[0]))
        trainAUC2.append(float(parsed[1]))
        trainLoss.append(float(parsed[2]))
        validAUC1.append(float(parsed[3]))
        validAUC2.append(float(parsed[4]))
        validLoss.append(float(parsed[5]))
    if not line: break
f.close()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle('asdf')

ax1.set_ylim(0.3, 0.7)
ax1.plot(trainAUC1, label = 'train auc 1')
ax1.plot(trainAUC1, label = 'train auc 1')
ax1.plot(validAUC1, label = 'validation auc 1')
ax1.plot(validAUC2, label = 'validation auc 1')
ax1.legend()
ax1.set_title('AUC')

ax2.set_ylim(min(min(trainLoss), min(validLoss))-0.1, max(max(trainLoss), max(validLoss)+0.1))
ax2.plot(trainLoss, label = 'train loss')
ax2.plot(validLoss, label = 'validation loss')
ax2.legend()
ax2.set_title('Loss')

fig.savefig(path + 'asdf' '.png')
```
